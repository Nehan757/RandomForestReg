```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import mplcyberpunk
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.ensemble import RandomForestRegressor  
reg = RandomForestRegressor(n_estimators=10 , random_state=0) 

# essemble is a techniqye in which multiple models are combined into 1 to improve the performance 
#specifies the number of trees in the random forest. The default value is 100

reg.fit(x,y)
 
```

```python
reg.predict([[6.5]])
```

```python
plt.style.use('cyberpunk')
plt.scatter(x,y,color='gold')
mplcyberpunk.make_scatter_glow

plt.plot(x,reg.predict(x))
mplcyberpunk.make_lines_glow

plt.show()
```

```python
xg = np.arange(min(x),max(x),0.01)
xg = xg.reshape(len(xg),1)

plt.scatter(x,y,color="gold")
mplcyberpunk.make_scatter_glow
plt.scatter(6.5,reg.predict([[6.5]]),color='red')

plt.plot(xg,reg.predict(xg))


```
